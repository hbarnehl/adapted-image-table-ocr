```python
import pandas as pd
```

```python
folder = '/home/hennes/Internship/constituencies'
df = pd.read_csv(folder+'/AC001.csv')
```

Moving columns to right when they end with nan

```python
# while loop make sure that repeated until no more nan on right side
while True in df.iloc[:,-1:].isna().all(axis=1).unique():
    mask = df.iloc[:,-1:].isna().all(axis=1)
    df.loc[mask, :] = df.loc[mask, :].shift(1, axis=1)

# when all nan are on left side, we can turn them to 0
df = df.fillna(0)
```

```python
repl_dict = {'$':'5',
            '(4)':'(A)',
            '4)':'(A)',
            'A':'(A)',
            'F':'0',
            'v':'0',
            '_':'',
            ']':'',
            '[':'',
            '|':''} 
```

```python
df = df.replace(None, repl_dict, regex=True)
```
